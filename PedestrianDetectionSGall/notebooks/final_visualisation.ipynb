```python
import pandas as pd

import matplotlib.pyplot as plt 

import csv
```

```python
df = pd.read_csv('final_results.csv')
```

```python
print(df.head(10))
```

```python
print(df.info())
```

```python
df.plot('Simulations', 'SSE X')
```

```python
df_0 = df[df['Pedestrian'] == 0]

df_1 = df[df['Pedestrian'] == 1]

df_2 = df[df['Pedestrian'] == 2]
```

```python
print(df_0.head)
```

```python
df_0.plot('Simulations', 'SSE X')

df_1.plot('Simulations', 'SSE X')

df_2.plot('Simulations', 'SSE X')
```

```python
df_0.plot('Simulations', 'SSE Y')

df_1.plot('Simulations', 'SSE Y')

df_2.plot('Simulations', 'SSE Y')
```

```python
plt.plot(df_0['Simulations'], df_0["Real X"], color = 'k')
plt.plot(df_0['Simulations'], df_0["Predicted X"], color = 'r', linestyle = '-.', label = 'Predicted X')
plt.title("Pedestrian 0 X coordinate vs prediciton")
plt.xlabel("Simulation")
plt.ylabel("X Coordinate Range")
plt.show()
```

```python
plt.plot(df_0['Simulations'], df_0["Real Y"], color = 'k')
plt.plot(df_0['Simulations'], df_0["Predicted Y"], color = 'b', linestyle = '-.', label = 'Predicted Y')
plt.title("Pedestrian 0 Y coordinate vs prediciton")
plt.xlabel("Simulation")
plt.ylabel("Y Coordinate Range")
plt.show()
```

```python
plt.plot(df_1['Simulations'], df_1["Real X"], color = 'k')
plt.plot(df_1['Simulations'], df_1["Predicted X"], color = 'r', linestyle = '-.', label = 'Predicted X')
plt.title("Pedestrian 1 X coordinate vs prediciton")
plt.xlabel("Simulation")
plt.ylabel("X Coordinate Range")
plt.show()
```

```python
plt.plot(df_1['Simulations'], df_1["Real Y"], color = 'k')
plt.plot(df_1['Simulations'], df_1["Predicted Y"], color = 'b', linestyle = '-.', label = 'Predicted Y')
plt.title("Pedestrian 1 Y coordinate vs prediciton")
plt.xlabel("Simulation")
plt.ylabel("Y Coordinate Range")
plt.show()
```

```python
plt.plot(df_2['Simulations'], df_2["Real X"], color = 'k')
plt.plot(df_2['Simulations'], df_2["Predicted X"], color = 'r', linestyle = '-.', label = 'Predicted X')
plt.title("Pedestrian 2 X coordinate vs prediciton")
plt.xlabel("Simulation")
plt.ylabel("X Coordinate Range")
plt.show()
```

```python
plt.plot(df_2['Simulations'], df_2["Real Y"], color = 'k')
plt.plot(df_2['Simulations'], df_2["Predicted Y"], color = 'b', linestyle = '-.', label = 'Predicted Y')
plt.title("Pedestrian 2 Y coordinate vs prediciton")
plt.xlabel("Simulation")
plt.ylabel("Y Coordinate Range")
plt.show()
```
