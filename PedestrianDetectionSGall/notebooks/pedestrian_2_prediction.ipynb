```python
from scripts import memspectrum as me

import numpy as np

np.random.seed(0)

import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
# Making dataset from pedestrian 2 movments 

#782, 786, 784, 785, 783, 774, 785, 786, 794, 793, 789, 788, 781, 771
person_2_X_2 = [783, 774, 785, 786, 794, 793, 789, 788, 781, 771]

# 1452, 1447, 1440, 1433, 1426, 1422, 1402, 1392, 1382, 1378, 1367, 1358, 1353, 1342
person_2_Y_2 = [1426, 1422, 1402, 1392, 1382, 1378, 1367, 1358, 1353, 1342]

# 782, 785, 774, 790, 789, 775, 766, 769, 760, 750, 753, 751, 739, 733
person_2_X_5 = [789, 775, 766, 769, 760, 750, 753, 751, 739, 733]

# 1452, 1436, 1422, 1385, 1367, 1347, 1332, 1319, 1305, 1289, 1271, 1261, 1246, 1234
person_2_Y_5 = [1367, 1347, 1332, 1319, 1305, 1289, 1271, 1261, 1246, 1234]

# 0, 0, 782, 774, 789, 766, 760, 753, 739, 739, 725, 729, 716, 712
person_2_X_10 = [789, 766, 760, 753, 739, 739, 725, 729, 716, 712]

# 0, 0, 1452, 1422, 1367, 1332, 1305, 1271, 1246, 1228, 1212, 1186, 1168, 1160
person_2_Y_10 = [1367, 1332, 1305, 1271, 1246, 1228, 1212, 1186, 1168, 1160]
```

```python
N, dt = 10, 1  # Number of samples and sampling interval
no_pred = 2
sims = 1000
time = np.arange(0, N) * dt
```

```python
plt.plot(time[:-1], person_2_X_2[:-1], color = 'k')
plt.title("Person 1 X coordinate")
plt.xlabel("Frame intervals of 2")
plt.ylabel("The X coordinate of Pedestrian Centroid")
plt.show()

plt.plot(time[:-1], person_2_Y_2[:-1], color = 'k')
plt.title("Person 1 Y coordinate")
plt.xlabel("Frame intervals of 2")
plt.ylabel("The Y coordinate of Pedestrian Centroid")
plt.show()

plt.plot(time[:-1], person_2_X_5[:-1], color = 'k')
plt.title("Person 1 X coordinate")
plt.xlabel("Frame intervals of 5")
plt.ylabel("The X coordinate of Pedestrian Centroid")
plt.show()

plt.plot(time[:-1], person_2_Y_5[:-1], color = 'k')
plt.title("Person 1 Y coordinate")
plt.xlabel("Frame intervals of 5")
plt.ylabel("The Y coordinate of Pedestrian Centroid")
plt.show()

plt.plot(time[:-1], person_2_X_10[:-1], color = 'k')
plt.title("Person 1 X coordinate")
plt.xlabel("Frame intervals of 10")
plt.ylabel("The X coordinate of Pedestrian Centroid")
plt.show()

plt.plot(time[:-1], person_2_Y_10[:-1], color = 'k')
plt.title("Person 1 Y coordinate")
plt.xlabel("Frame intervals of 10")
plt.ylabel("The Y coordinate of Pedestrian Centroid")
plt.show()
```

```python
forecast = me.get_forecast(person_2_X_2[:-no_pred], no_pred, sims)
median_X_2 = np.median(forecast, axis = 0) # Ensemble median
average_X_2 = np.average(forecast, axis=0)
```

```python
forecast = me.get_forecast(person_2_Y_2[:-no_pred], no_pred, sims)
median_Y_2 = np.median(forecast, axis = 0) # Ensemble median
average_Y_2 = np.average(forecast, axis=0)
```

```python
plt.plot(time, person_2_X_2, color = 'k')
plt.plot(time[N-no_pred:], person_2_X_2[N-no_pred:], color = 'g', linestyle = '-.', label = 'Observed data')
plt.plot(time[N-no_pred:], median_X_2, color = 'r', label = 'median estimate')
plt.plot(time[N-no_pred:], average_X_2, color = 'b', label = 'average estimate')
plt.title("Person 1 X coordinate prediction")
plt.xlabel("Frame intervals of 10")
plt.ylabel("The X coordinate of Pedestrian Centroid")
plt.show()
```

```python
plt.plot(time, person_2_Y_2, color = 'k')
plt.plot(time[N-no_pred:], person_2_Y_2[N-no_pred:], color = 'g', linestyle = '-.', label = 'Observed data')
plt.plot(time[N-no_pred:], median_Y_2, color = 'r', label = 'median estimate')
plt.plot(time[N-no_pred:], average_Y_2, color = 'b', label = 'average estimate')
plt.title("Person 1 Y coordinate prediction")
plt.xlabel("Frame intervals of 10")
plt.ylabel("The Y coordinate of Pedestrian Centroid")
plt.show()
```

```python
real_X_values = person_2_X_2[N-no_pred:]
print("Real X Values = ", real_X_values)
print("Median X Values = ", median_X_2)
print("Average X Value = ", average_X_2)
SSE_X = np.sum((real_X_values - median_X_2)**2)
SSR_X = np.sum((real_X_values - average_X_2)**2)
SST_X = np.sum(SSR_X + SSE_X)
print("SSE X = ", SSE_X)
print("SSR X = ", SSR_X)
print("SST X = ", SST_X)
```

```python
# Sum Squared Error just as a quick evaluation
real_Y_values = person_2_Y_2[N-no_pred:]
print("Real Y Values = ", real_Y_values)
print("Median Y Values = ", median_Y_2)
print("Average Y Value = ", average_Y_2)
SSE_Y = np.sum((real_Y_values - median_Y_2)**2)
SSR_Y = np.sum((real_Y_values - average_Y_2)**2)
SST_Y = np.sum(SSR_Y + SSE_Y)
print("SSE Y = ", SSE_Y)
print("SSR Y = ", SSR_Y)
print("SST Y = ", SST_Y)
```
